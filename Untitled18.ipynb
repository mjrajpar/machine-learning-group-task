```python
prem_m = pd.read_csv('premiermatches.csv')
print(prem_m.head())
```

```python
# Check the shape of the dataset
# Provides an overview of the dataset
print("Dataset shape:", prem_m.shape)

# View value counts for the "team" column
# Displays the distribution of prem_m per team
print("\nValue counts for 'team':")
print(prem_m["team"].value_counts())

# Filter prem_m for a specific team (e.g., Liverpool) for exploration
liverpool_prem_m = prem_m[prem_m["team"] == "Liverpool"]
print("\nSample match for Liverpool:")
print(liverpool_prem_m.head())


# Check data types of columns
# Helps determine the preprocessing steps required
print("\nColumn data types:")
print(prem_m.dtypes)

```

```python
# Create a numerical encoding for the "venue" column
# Converts categorical variables into numerical format for model compatibility
prem_m["venue_code"] = prem_m["venue"].astype("category").cat.codes

# Create a numerical encoding for the "opponent" column
prem_m["opp_code"] = prem_m["opponent"].astype("category").cat.codes

# Extract the hour from the "time" column
# Captures patterns based on the time of the match

prem_m["hour"] = pd.to_numeric(prem_m["time"].str.replace(":.+", "", regex=True), errors='coerce').astype('Int64')

# Convert "date" column to a datetime object and extract the day of the week
# Captures the potential effects of weekday/weekend prem_m
prem_m["date"] = pd.to_datetime(prem_m["date"], format="%d/%m/%Y", errors="coerce", dayfirst=True)
prem_m["day_code"] = prem_m["date"].dt.dayofweek
prem_m["date"] = pd.to_datetime(prem_m["date"], errors="coerce")
print(prem_m.dtypes)
```

```python
# Create a single target column for multi-class classification
# Encode "result" into numerical categories: 0 = Loss, 1 = Draw, 2 = Win
# Define result mapping
result_mapping = {"L": -1, "W": 3,"D": 0}

# Map and handle potential issues
prem_m["target"] = prem_m["result"].map(result_mapping).fillna(-1).astype(int)

# Verify the changes
print(prem_m.head())

```

```python
# Drop unnecessary columns
columns_to_drop = ['Unnamed: 0', 'comp', 'round',
                   'match report']
prem_m = prem_m.drop(columns=columns_to_drop)

# Separate features and target
X = prem_m.drop(columns=["target"])  # Features
y = prem_m["target"]  # Target (0 = Loss, 1 = Draw, 2 = Win)

# Verify the changes
print(prem_m.head())
```

```python
# Check for missing values

print("\nMissing values in dataset:")
print(prem_m.isnull().sum())

# Drop rows with missing values
# prem_m = prem_m.dropna()

# Split data into training and testing sets using train_test_split
from sklearn.model_selection import train_test_split # Importing the necessary function
train, test = train_test_split(prem_m, test_size=0.2, random_state=42) # Creating train and test sets


print(f"Training set size: {train.shape}")
print(f"Testing set size: {test.shape}")

# Check for missing values
print("\nMissing values in dataset:")
print(prem_m.isnull().sum())

```

```python
# Define predictors
predictors = ["venue_code", "opp_code", "hour", "day_code"]

# Initialize and train Random Forest model
rfm = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=2)


# Check predictors and target column
print("Predictors used for training:", predictors)
print("Columns in train:", train.columns)
```

```python



rfm.fit(train[predictors], train["target"])

# Predict on the test set
preds = rfm.predict(test[predictors])


print("\nAccuracy Score:")
error = accuracy_score(test["target"], preds)
print(error)

print("\nConfusion Matrix:")
print(confusion_matrix(test["target"], preds))

print("\nClassification Report:")
print(classification_report(test["target"], preds, target_names=["Loss", "Draw", "Win"]))

# Feature Importance
print("\nFeature Importances:")
feature_importances = pd.Series(rfm.feature_importances_, index=predictors)
print(feature_importances.sort_values(ascending=False))
```

```python
print("\nPrecision Score:")
precision_score(test["target"], preds, average='weighted') # Add the 'average' parameter with a suitable option like 'weighted'
```

```python
grouped_prem_m = prem_m.groupby("team")
group = grouped_prem_m.get_group("Manchester City").sort_values("date")
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group

cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
# Define predictors
predictors = ["venue_code", "opp_code", "hour", "day_code"]

# Create an imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform it
train_imputed = imputer.fit_transform(train[predictors])

# Convert the imputed data back to a pandas DataFrame
train_imputed = pd.DataFrame(train_imputed, columns=predictors, index=train.index)

# Initialize Logistic Regression model
log_reg = LogisticRegression(max_iter=500, random_state=42)

# Train Logistic Regression model using the imputed data
log_reg.fit(train_imputed, train["target"])

# Now, you need to impute the missing values in the test set as well
test_imputed = imputer.transform(test[predictors])  # Use transform, not fit_transform
test_imputed = pd.DataFrame(test_imputed, columns=predictors, index=test.index)

# Predict on the test set using Logistic Regression
log_preds = log_reg.predict(test_imputed)
```

```python
# Evaluate the Logistic Regression model
print("\nLogistic Regression Accuracy Score:")
log_error = accuracy_score(test["target"], log_preds)
print(log_error)

print("\nLogistic Regression Confusion Matrix:")
print(confusion_matrix(test["target"], log_preds))

print("\nLogistic Regression Classification Report:")
print(classification_report(test["target"], log_preds, target_names=["Loss", "Draw", "Win"],))


# Compare with Random Forest Model
print("\nModel Comparison:")
print(f"Random Forest Accuracy: {error}")
print(f"Logistic Regression Accuracy: {log_error}")
```
